```python
import librosa
from scipy.io.wavfile import read
from librosa.feature.rhythm import tempo
from librosa.onset import onset_strength
import librosa.display
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import Audio
```

```python
y, sr = librosa.load('beatles.wav', duration = 30, sr = 44100) # 30s audio

tempo2, actual_beats = librosa.beat.beat_track(y = y, sr= sr, hop_length=64, units="time")
print("Tempo:",tempo2)
beats = np.arange(0,30,60/np.round(tempo2))
print("Beats:",beats)
Audio(y, rate = sr)
```

```python
plt.figure(figsize=(18,10))
# librosa.display.waveshow(y=y, sr=sr)
plt.plot(beats, np.zeros(len(beats)), '*')
```
